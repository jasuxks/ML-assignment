```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('iris.csv')
```

```python
sns.pairplot(data, hue='Species')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
for species in data['Species'].unique():
    sns.histplot(data[data['Species'] == species]['SepalLengthCm'], kde=True, label=species)
plt.legend()
plt.title('Sepal Length Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='Species', y='PetalWidthCm', data=data)
plt.title('Petal Width by Species')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.violinplot(x='Species', y='PetalLengthCm', data=data)
plt.title('Petal Length by Species')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.swarmplot(x='Species', y='SepalWidthCm', data=data)
plt.title('Sepal Width by Species')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data['Species'])
plt.title('Species Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', data=data, hue='Species')
plt.title('Sepal Length vs. Sepal Width')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x='Id', y='PetalLengthCm', data=data, hue='Species')
plt.title('Petal Length by Id')
plt.show()
```

```python
sns.pairplot(data, hue='Species', diag_kind='kde')
plt.show()
```

```python
sns.pairplot(data, hue='Species', kind='reg')
plt.show()
```
